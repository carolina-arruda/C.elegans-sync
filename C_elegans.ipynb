# Using crosstab to plot data

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import rcParams
rcParams['figure.figsize'] = 10,10
```

```python
elegans = pd.read_csv('Data-EJ248.csv')
```

```python
elegans.shape
```

```python
elegans.head()
```

```python
elegans.columns = ['Neuron', 'Index', 'Localization', '3_mod', 'Class', 'Ganglion', '5_mod', '10_mod']
```

```python
elegans.head()
```

```python
elegans.info()
```

```python
elegans.Localization.value_counts()
```

```python
elegans.Class.value_counts()
```

```python
elegans.Ganglion.value_counts()
```

```python
ct11 = pd.crosstab(elegans.Ganglion, elegans.Class, normalize='index')
ct11.plot.bar(stacked = True, rot = 0, fontsize = 20)
plt.title('Type of classes in each ganglia', fontsize = 20)
plt.ylabel('% of total', fontsize = 20)
plt.xlabel('')
plt.show()
```
